```python
# Cell 1: Imports
import pandas as pd
import numpy as np
import json
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
print("Imports completed successfully.")
```

```python
# Cell 2: Load Netflix Dataset
netflix_df = pd.read_csv('../data/raw/netflix_titles.csv')

print("Netflix Dataset Shape:", netflix_df.shape)
print("\nColumn Names:")
print(netflix_df.columns.tolist())
print("\nFirst few rows:")
netflix_df.head()
```

```python
# Cell 3: Analyze Netflix Data
print("Data Info:")
print(netflix_df.info())
print("\n" + "="*50)
print("Missing Values:")
print(netflix_df.isnull().sum())
print("\n" + "="*50)
print("Content Types:")
print(netflix_df['type'].value_counts())
```

```python
# Cell 4: Load TMDB Movies Dataset
tmdb_movies = pd.read_csv('../data/raw/tmdb_5000_movies.csv')

print("TMDB Movies Shape:", tmdb_movies.shape)
print("\nColumn Names:")
print(tmdb_movies.columns.tolist())
print("\nFirst few rows:")
tmdb_movies.head()
```

```python
# Cell 5: Analyze TMDB Data
print("Data Info:")
print(tmdb_movies.info())
print("\n" + "="*50)
print("Missing Values:")
print(tmdb_movies.isnull().sum())
print("\n" + "="*50)
print("Vote Average Distribution:")
print(tmdb_movies['vote_average'].describe())
```

```python
# Cell 6: Load TMDB Credits
tmdb_credits = pd.read_csv('../data/raw/tmdb_5000_credits.csv')

print("TMDB Credits Shape:", tmdb_credits.shape)
print("\nColumn Names:")
print(tmdb_credits.columns.tolist())
tmdb_credits.head()
```

```python
# Cell 7: Sample Analysis - Top Genres in Netflix
# Clean and split genres
netflix_df['listed_in_clean'] = netflix_df['listed_in'].fillna('Unknown')
all_genres = netflix_df['listed_in_clean'].str.split(', ').explode()
print("Top 15 Genres on Netflix:")
print(all_genres.value_counts().head(15))
```

```python
# Cell 9: Identify High-Quality Content for Documents
# Movies with good ratings and descriptions
quality_movies = tmdb_movies[
    (tmdb_movies['vote_average'] >= 7.0) & 
    (tmdb_movies['vote_count'] >= 1000) &
    (tmdb_movies['overview'].notna())
].copy()

print(f"High Quality Movies Found: {len(quality_movies)}")
print("\nTop 10 by Rating:")
print(quality_movies.nlargest(10, 'vote_average')[['title', 'vote_average', 'vote_count']])
```

```python
# Cell 10: Save analysis results
quality_movies.to_csv('../data/processed/quality_movies.csv', index=False)
print("Saved quality movies to data/processed/quality_movies.csv")
```
